```
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit
spark = SparkSession.builder.appName("selectWithColumnFilter").getOrCreate()
df = spark.read.options(interSchema=True,header=True).csv("/FileStore/tables/StudentData.csv").withColumn("roll",col("roll").cast("String"))
df.show()
```

```
df = df.withColumn("total_marks", lit(120))
df.show()
```

```
df = df.withColumn("average_marks", (df.marks/df.total_marks)*100)
df.show()
```

```
oop_over_eighty = df.filter((df.course == 'OOP') & (df.marks > 80))
oop_over_eighty.show()
```

```
cloud_over_sixty = df.filter((df.course == 'Cloud') & (df.marks > 60))
cloud_over_sixty.show()
```

```
print(oop_over_eighty.select("name").show())
```

```
print(cloud_over_sixty.select("name").show())
```
